Since the dataset contains string values, we must correct those that might be misspelled, that's why we need to see the columns and select the information we would like to correct by using the code:

```
df["column"].unique()
```



````python
df.columns
````

The code provides a view of the string values that are contained in the column.

````python
df["Campaign_Name"].unique()
````

````python
df["Campaign_Name"] = df["Campaign_Name"].replace({
    "DataAnalyticsCourse": "Data Analytics Course",
    "Data Analytics Corse": "Data Analytics Course",
    "Data Analytcis Course": "Data Analytics Course",
    "Data Anlytics Corse": "Data Analytics Course"

})
````

````python
df["Location"].unique()
````

````python
df["Location"] = df["Location"].replace ({
    "hyderabad" : "Hyderabad",
    "Hyderbad" : "Hyderabad" ,
    "hydrebad" : "Hyderabad",
    "HYDERABAD" : "Hyderabad"
})
````

````python
df["Device"].unique()
````

````python
df["Device"] = df["Device"].replace ({
    "desktop" : "Desktop",
    "DESKTOP" : "Desktop" ,
    "tablet" : "Tablet",
    "TABLET" : "Tablet",
    "MOBILE" : "Mobile",
    "mobile" : "Mobile"

})
````

Another thing we must check is the incorrect format in the date. The code below presents the different amount of dates that appear in the dataset:

````python
df["Ad_Date"].value_counts()
````

Now, you correct it by using the code below:

````python
df["Ad_Date"] = pd.to_datetime(df["Ad_Date"], format="mixed", dayfirst=True)
````

````python
df["Ad_Date"].value_counts()
````

The next thing we should do is to look for the null values. This code helps us with that:


````python
df.isnull().sum()
````

Now, we fill them:

````python
df.fillna("N/A", inplace=True)
````

````python
df.to_csv("cleaned_dataset.csv", index=False)
````
